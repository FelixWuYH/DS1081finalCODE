```python
#DS1ex6_10720129張書維
class  Title :
    def __init__(self,snum,sname,dnum,dname,Type,level,city,system,nstd,nprof,ngrad) :
        self.snum = snum  # // school number , name // department number
        self.sname = sname
        self.dnum = dnum   
        self.dname = dname
        self.Type = Type  # // 日間/進修別  /等級別 
        self.level= level 
        self.city= city
        self.system = system
        self.nstd = nstd        # 學生人數 教師 畢業學生人數
        self.nprof = nprof
        self.ngrad = ngrad
        self.left = None
        self.right = None
        self.parent = None
        return 
```

```python
def Max(a,b) :
    if a > b : return a 
    else : return b 
```

```python
class BST :
    def __init__(self) :
        self.gradroot = None   # 人數 root 
        self.nameroot = None   # 學校 root
    def Addbygrad( self,NewNode) :  #　以人數為鍵直加入資料
        pre = None
        walk = self.gradroot
        New = NewNode
        while walk != None :       # 找到新節點該放的位置
            pre = walk
            if New.ngrad < walk.ngrad : walk = walk.left
            else : walk = walk.right 
        New.parent = pre   # 設定parent
        if pre == None : self.gradroot = New      #放入
        elif  New.ngrad < pre.ngrad : pre.left = New
        else : pre.right = New
        return self.gradroot
    def Addbyname( self,NewNode) :    #　以學校名稱為鍵直加入資料
        pre = None
        walk = self.nameroot
        New = NewNode
        while walk != None :
            pre = walk 
            if New.sname < walk.sname : walk = walk.left
            else : walk = walk.right
        New.parent = pre 
        if pre == None : self.nameroot = New 
        elif  New.sname < pre.sname : pre.left = New
        else : pre.right = New
        return self.nameroot
    def Height(self,root) :# 樹高計算
        if root == None : return 0
        left = self.Height(root.left)
        right = self.Height(root.right)
        return 1 + Max(left,right)
    def RecurSearchbygrad(self,i,inputgrad,cur) :
        stop = False 
        if cur != None :
            if  cur.ngrad >= inputgrad :
                i[0]+=1
                print(i,"\t", cur.sname,"\t",cur.dname, "\t",cur.Type,"\t",cur.level,"\t",cur.nstd,"\t",cur.nprof,"\t",cur.ngrad )
            if cur.ngrad < inputgrad : # 設立停止條件
                if cur.left != None : stop = True 
            if stop == False : self.RecurSearchbygrad(i,inputgrad,cur.left)   # 若達成條件就停止
            self.RecurSearchbygrad(i,inputgrad,cur.right) 
    def RecurSearchbyname(self,i,inputname,cur) :
        lstop = rstop =False 
        if cur != None :
            if  cur.sname == inputname :
                i[0]+=1
                print(i,"\t", cur.sname,"\t",cur.dname, "\t",cur.Type,"\t",cur.level,"\t",cur.nstd,"\t",cur.nprof,"\t",cur.ngrad )
            if  cur.sname < inputname: # 設立停止條件
                if cur.left != None :  lstop = True 
            if  cur.sname > inputname: # 設立停止條件
                if cur.right != None : rstop = True 
            if  lstop==False : self.RecurSearchbyname(i,inputname,cur.left)  # 若達成條件就停止
            if  rstop==False : self.RecurSearchbyname(i,inputname,cur.right)  # 若達成條件就停止
    def Searchbygrad(self) :
        i=[0]
        inputgrad = input( "Input the number of graduates: " )  
        cur = self.gradroot  
        self.RecurSearchbygrad(i,int(inputgrad),cur) 
    def Searchbyname(self) :
        i=[0]
        inputname = input( "Input the School Name: " )  
        cur = self.nameroot
        self.RecurSearchbyname(i,inputname,cur) 
    def DelSearchbyname(self,name) :  # 找到相同學校名稱的節點
        walk = self.nameroot 
        while  walk != None and name != walk.sname :
            if  name > walk.sname : walk = walk.right 
            else :walk = walk.left 
        return walk 
    def CheckNode(self,Node,walk) :  # 當畢業人數相同時確認學校名稱是否相同
        if  Node.sname != walk.sname :return False
        return True 
    def DelSearchbygrad(self,Node):  # 回傳刪除的節點
        walk = self.gradroot   
        while  walk != None :
            if walk.ngrad == Node.ngrad :
                if self.CheckNode(Node,walk) == True : 
                    break 
            if walk.ngrad > Node.ngrad: walk = walk.left 
            else : walk = walk.right
        return walk 
            
    def Deletegradbyname(self,delete) :
        delgrad = self.DelSearchbygrad(delete)
            
        change =None
        if  delgrad.left != None and delgrad.right !=None:
            change = self.Leftmost(delgrad.right) 
            if change.right != None :
                if  change.parent != delgrad :
                    change.right.parent = change.parent 
                    change.parent.left = change.right 
                else :
                    change.right.parent = change.parent 
                    change.parent.right = change.right 
            else :
                if change.parent.right == change: change.parent.right = None 
                elif change.parent.left == change:  change.parent.left = None
            self.PassValue(change,delgrad) 

            change = None
        elif delgrad.left == None and delgrad.right == None  :
            if  delgrad.parent != None :  
                if delgrad.parent.left == delgrad: delgrad.parent.left = None  
                elif delgrad.parent.right == delgrad : delgrad.parent.right = None  
            else : self.gradroot = None
        elif  delgrad.left == None and delgrad.right != None  :
            change = self.Leftmost(delgrad.right)
            if  change.right != None   :
                if  change.parent !=delgrad :
                    change.right.parent = change.parent 
                    change.parent.left = change.right 
                else :
                    change.right.parent = change.parent
                    change.parent.right = change.right
            else :
                if  change.parent.right == change:change.parent.right = None
                elif  change.parent.left == change:change.parent.left = None  

            self.PassValue(change,delgrad) 
            change = None
        elif  delgrad.left != None and delgrad.right == None:
            change = self.Rightmost(delgrad.left) 
            if  change.left != None :
                if  change.parent !=delgrad :
                    change.left.parent = change.parent 
                    change.parent.right = change.left
                else :
                    change.left.parent = change.parent 
                    change.parent.left = change.left
            else :
                if  change.parent.right == change:change.parent.right = None
                elif  change.parent.left == change:change.parent.left = None
            self.PassValue(change,delgrad) 
            change = None
    def Leftmost(self,cur) :
        while  cur.left!= None  : cur = cur.left 
        return cur 
    def Rightmost(self,cur) :
        while  cur.right!= None  : cur = cur.right
        return cur 
    def PassValue(self,change,New) :
        New.dname = change.dname 
        New.sname = change.sname 
        New.Type =  change.Type 
        New.level =  change.level 
        New.nstd =  change.nstd 
        New.nprof = change.nprof 
        New.ngrad =  change.ngrad 
        
    def Delete(self,name) :      
        delete = self.DelSearchbyname(name) # 找到要刪除的節點 
        i = 0 
        if delete == None : print("NOT FOUND")
        while delete != None:
            print("[",i+1,"]\t", delete.sname,"\t",delete.dname, "\t",delete.Type,"\t",delete.level,"\t",delete.nstd,"\t",delete.nprof,"\t",delete.ngrad )
            self.Deletegradbyname(delete) # 將節點傳入 並在畢業人數樹找到對應節點並刪除
            i+=1
            change =None
            if  delete.left != None and delete.right !=None:  # 被刪除節點 有左右小孩
                change = self.Leftmost(delete.right)  # 讓 change 去替代被刪除節點的位置 右子樹的最左端鑑直 即為最接近被刪除節點
                if change.right != None :             # 切斷change 前後鏈結 並接上
                    if  change.parent != delete :
                        change.right.parent = change.parent 
                        change.parent.left = change.right 
                    else :
                        change.right.parent = change.parent 
                        change.parent.right = change.right 
                else :
                    if change.parent.right == change: change.parent.right = None 
                    elif change.parent.left == change:  change.parent.left = None
                self.PassValue(change,delete) # 將 change 節點的直傳給被刪除之節點
                change = None  # 節點被刪除
            elif delete.left == None and delete.right == None  : # 被刪除節點無小孩
                if  delete.parent != None :  # 可以直接刪掉
                    if delete.parent.left == delete: delete.parent.left = None  
                    elif delete.parent.right == delete : delete.parent.right = None  
                    delete = None
                else : self.nameroot == None 
            elif  delete.left == None and delete.right != None  :# 被刪除節點有右小孩
                change = self.Leftmost(delete.right)
                if  change.right != None   :
                    if  change.parent != delete :
                        change.right.parent = change.parent 
                        change.parent.left = change.right 
                    else :
                        change.right.parent = change.parent 
                        change.parent.right = change.right
                else :
                    if  change.parent.right == change:change.parent.right = None
                    elif  change.parent.left == change:change.parent.left = None  
                self.PassValue(change,delete) 
                change = None
                
            elif  delete.left != None and delete.right == None:
                change = self.Rightmost(delete.left) 
                if  change.left != None :
                    if  change.parent !=delete :
                        change.left.parent = change.parent 
                        change.parent.right = change.left
                    else :
                        change.left.parent = change.parent 
                        change.parent.left = change.left
                else :
                    if  change.parent.right == change:change.parent.right = None
                    elif  change.parent.left == change:change.parent.left = None

                self.PassValue(change,delete) 
                change = None
            delete = self.DelSearchbyname(name) 
```

```python
def CreateNode(List,i) :
    New= Title(List[i].snum,List[i].sname,List[i].dnum,List[i].dname,List[i].Type,List[i].level,List[i].city,List[i].system,List[i].nstd,List[i].nprof,List[i].ngrad )
    return New
```

```python
 def CreateTree(tree,List):  # 建立樹
     i= 0
     while i < len(List) :  
         tree.gradroot = tree.Addbygrad(CreateNode(List,i))
         tree.nameroot = tree.Addbyname(CreateNode(List,i))
         i+=1
     print("Tree heights:\n{Number of graduates} = ",tree.Height(tree.gradroot) )
     print( "{School name} = ",tree.Height(tree.nameroot))
     return tree
```
