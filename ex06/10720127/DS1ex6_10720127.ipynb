```python
class node: #資料型態
    def __init__(self, schoolname, majorname, time, level, studentNum, teacherNum, graduated):
        self.schoolname =schoolname
        self.majorname =majorname
        self.time =time
        self.level =level
        self.studentNum =studentNum
        self.teacherNum =teacherNum
        self.graduated =graduated
        self.left = None
        self.right = None
    
    def insertG(self, schoolname, majorname, time, level, studentNum, teacherNum, graduated):
        if self.graduated <= graduated :    #如果當前資料大於樹的則往右邊走
            if self.left != None :  #如果右邊小孩有資料則再跑一次遞迴
                self.left.insertG(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
            else :  #右邊小孩沒東西的話，就直接插入
                self.left = node(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
        else :    #如果當前資料小於樹的則往左邊走
            if self.right :
                self.right.insertG(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
            else :
                self.right = node(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
                
    def insertN(self, schoolname, majorname, time, level, studentNum, teacherNum, graduated):
        if self.schoolname <= schoolname :  #如果當前資料大於樹的則往右邊走
            if self.left != None:  #如果右邊小孩有資料則再跑一次遞迴
                self.left.insertN(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
            else :    #右邊小孩沒東西的話，就直接插入
                self.left = node(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
        else :    #如果當前資料小於樹的則往左邊走
            if self.right :
                self.right.insertN(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
            else :
                self.right = node(schoolname, majorname, time, level, studentNum, teacherNum, graduated)
            
    
        
```

```python
class Sorting :
    def __init__(self) : 
        self.test = []  #存一開始的資料
        self.grad = None  # grad的樹根
        self.name = None  # name的樹根
        
    def PrintTree(self, test): #印出全部資料
        print("學校名稱\t科系名稱\t日間∕進修別\t等級別\t學生數\t教師數\t上學年度畢業生數")
        i = 0
        while i < len(test) :
            print(test[i].schoolname,'\t',test[i].majorname,'\t',test[i].time,'\t',test[i].level,'\t',test[i].studentNum,'\t',test[i].teacherNum,'\t',test[i].graduated)
            i += 1
    
    def SortGrad( self, temp ) :
        if self.grad != None:  #如果root不是NULL就丟進去insertG做遞迴呼叫插入
            self.grad.insertG(temp.schoolname, temp.majorname, temp.time, temp.level, temp.studentNum, temp.teacherNum, temp.graduated)
        else :  #root是NULL的話則直接插入
            self.grad = node(temp.schoolname, temp.majorname, temp.time, temp.level, temp.studentNum, temp.teacherNum, temp.graduated)
            
    def SortName( self, temp ) :
        if self.name != None :  #如果root不是NULL就丟進去insertN做遞迴呼叫插入
            self.name.insertN(temp.schoolname, temp.majorname, temp.time, temp.level, temp.studentNum, temp.teacherNum, temp.graduated)
        else : #root是NULL的話則直接插入
            self.name = node(temp.schoolname, temp.majorname, temp.time, temp.level, temp.studentNum, temp.teacherNum, temp.graduated)
        
               
            
        
    def Height( self, tree ) :  #用遞迴方式測試樹高
        if tree != None :
            leftHeight = self.Height( tree.left )
            rightHeight = self.Height( tree.right )
            if leftHeight > rightHeight :  #看哪一邊比較高就return他
                return leftHeight + 1   #每次遞迴結束+1樹高
            else :
                return rightHeight + 1
        else :
            return 0
```
