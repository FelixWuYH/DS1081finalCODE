```python
# 工作資訊
# arrival + duration <= timeout
class JobType():
    def __init__( self ):
        self.OID = 0 ; # order identifier
        self.arrival = 0 ; # arrival time
        self.duration = 0 ; # job duration
        self.timeout = 0 ; # expire time
```

```python
class JobList():
    # constructor
    def __init__( self ):
        self.aList = []
        self.aList.clear()
        self.fileID = "output" + fileTitleNum + ".txt"
        
        
    # 印出一個job
    def ShowJob( self, job ):
        print( job.OID , "\t" , job.arrival , "\t" , job.duration , "\t" , job.timeout )
        
        
        
    # 從檔案中取出資料
    def GetAll( self, fileName ):
        try: # { try
            file = open( fileName, 'r' ) 
            aLine = file.readline() # 先把第一行讀掉(四個item名稱)
            for aLine in file : # { 如果檔案還沒讀完 讀一行(含換行)
                
                aJob = JobType()
                value = 0 
                # 從字串頭到尾找一遍 跳過tab存進job
                for i in range( 0, len(aLine) ): # {
                    if ( '0' <= aLine[i] and aLine[i] <= '9' ): # 是數字就累加
                        value = value * 10 + ( int(aLine[i]) - int('0') )
                    else : # { 不是數字，則存到還沒存的欄位
                        if ( aJob.OID == 0 ):
                            aJob.OID = value
                        elif ( aJob.arrival == 0 ):
                            aJob.arrival = value
                        elif ( aJob.duration == 0 ):
                            aJob.duration = value
                        elif ( aJob.timeout == 0 ):
                            aJob.timeout = value
                        
                        value = 0 # 初始化為零
                    # } else
                # } for
                
                if ( aJob.timeout != 0 ) : # 如果最後一項不為零 則數據完整
                    self.aList.append(aJob) # 將此筆資料存進aList
                
            # } while
            
            file.close()
            return True
        # } try
        except:
            return False
    
    
    # display all jobs on screen
    def ShowAll( self ):
        print("OID\tArrival\tDuration\tTimeOut")
        for i in range( 0, len(self.aList) ):
            self.ShowJob( self.aList[i] )
            
    # check whether it is empty or not
    def IsEmpty( self ):
        if ( len(self.aList) == 0 ):
            return True;
        return False ;
    
    
    # get and remove first job
    def NextJob( self ):
        job = JobType()
        if ( self.IsEmpty() == False ):
            job = self.aList[0] # 先把第一項存下來
            self.aList.remove( self.aList[0] ) # 刪除第一項
        return job
    
```

```python
class JobQueue:
    # constructor
    def __init__( self, maxS ):
        self.cA = [] # circular array
        self.qFront = 0 # head of queue
        self.qBack = 0 # tail of queue
        self.qSize = 0 # current size of queue
        self.qMax = maxS # max size of queue
        for i in range(0, maxS):
            self.cA.append('0')
        
    # get the current queue length
    def Length( self ):
        return self.qSize
    
    # check whether it is empty
    def IsEmpty( self ):
        if ( self.qSize == 0 ):
            return True
        return False
    
    # check whether it is full
    def IsFull( self ):
        if ( self.qSize == self.qMax ):
            return True
        return False
    
    # append a new element
    def EnQueue( self, item ):
        if ( self.IsFull() == False ):
            if ( self.IsEmpty() ):
                self.qFront = self.qBack = 0 
            else:
                self.qBack = (self.qBack+1)%self.qMax
            
            self.cA[self.qBack] = item
            self.qSize = self.qSize + 1
            
            
    # get and drop first element
    def DeQueue( self ):
        if ( self.IsEmpty() == False ):
            item = self.cA[self.qFront]
            if ( self.qSize == 1 ):
                self.qFront = self.qBack = 0
            else:
                self.qFront = (self.qFront+1)%self.qMax
            self.qSize = self.qSize - 1
        return item
    
    
    # get the first element
    def GetFront( self ):
        if ( self.IsEmpty() == False ):
            return cA[self.qFront]
            
```

```python
class AnsList:
    
    class AbortType():
        def __init__( self ):
            self.OID = 0
            self.abort = 0
            self.delay = 0
    
    class DoneType():
        def __init__( self ):
            self.OID = 0
            self.departure = 0
            self.delay = 0
            
    # constructor
    def __init__( self ):
        self.abortJobs = [] # list of aborted jobs with three columns
        self.abortJobs.clear()
        self.doneJobs = [] # list of done jobs with three columns
        self.doneJobs.clear()
        self.avgDelay = 0.0 # average delay
        self.successRate = 0.0 # percentage of done jobs
        
        
    # display abort jobs on screen
    def __ShowAbortJobs( self ):
        print( "\t[Abort Jobs]\n\tOID\tAbort\tDelay")
        for i in range( 0, len(self.abortJobs) ):
            print( "[", i+1, "]", self.abortJobs[i].OID, "\t", self.abortJobs[i].abort, "\t", self.abortJobs[i].delay )
        
    # display done jobs on screen
    def __ShowDoneJobs( self ):
        print( "\t[Jobs Done]\n\tOID\tDeparture\tDelay")
        for i in range( 0, len(self.doneJobs) ):
            print( "[", i+1, "]", self.doneJobs[i].OID, "\t", self.doneJobs[i].departure, "\t", self.doneJobs[i].delay )
         
        
    # compute the statistics
    def __ComputeStat( self ):
        totalDelay = 0
        for i in range( 0, len(self.abortJobs) ):
            totalDelay = totalDelay + self.abortJobs[i].delay
        for i in range( 0, len(self.doneJobs) ):
            totalDelay = totalDelay + self.doneJobs[i].delay
            
        self.avgDelay = float( totalDelay / ( len(self.abortJobs) + len(self.doneJobs) ) )
        self.successRate = float( 100 * len(self.doneJobs) / ( len(self.abortJobs) + len(self.doneJobs) ) )
        
        
    # display statistics on screen
    def __ShowStat( self ):
        self.__ComputeStat()
        print( "[Average Delay]\t", self.avgDelay, " ms\n[Success Rate]\t", self.successRate, " %" )
        
        
    # add one aborted job
    def AddAbortJob( self, aJob, nowTime, fromQ ):
        # 以下口語化分為三種情況 
        # 1.前面還沒好就來，只能跟你說掰掰 
        # 2.排隊排到一半，輪到你時發現逾時了，掰掰 
        # 3.終於輪到我了，跑跑跑，啊 ！逾時了，掰掰  
        abortJob = self.AbortType()
        abortJob.OID = aJob.OID 
        if ( aJob.arrival < nowTime and fromQ == False ): # 現在(nowTime)才弄完，之前到達(arrival)的只好取消 
            abortJob.abort = aJob.arrival # 到達(arrival)就被告知取消(abort) 
        elif ( nowTime > aJob.timeout and fromQ ): # 從佇列取出時(nowTime)已經逾時(timeout)，只好取消 
            abortJob.abort = nowTime #被取消時間(abort)就是現在(nowTime) 
            nowTime = abortJob.abort 
        elif ( nowTime + aJob.duration > aJob.timeout ): # 執行到一半(nowTime)，發現已經逾時(timeout)，所以只好取消 
            abortJob.abort = aJob.timeout  # 取消時間(abort) == 發現已經逾時(timeout) 
            nowTime = abortJob.abort
        
        abortJob.delay = abortJob.abort - aJob.arrival # delay = 到達時間(arrival) - 取消時間(abort) 
        self.abortJobs.append( abortJob ) 
        return nowTime
    
    
    # add one done job
    def AddDoneJob( self, aJob, nowTime ):
        doneJob = self.DoneType()
        doneJob.OID = aJob.OID
        if ( nowTime < aJob.arrival ): # 如果aJob無需等待
            doneJob.departure = aJob.arrival + aJob.duration # 執行完時間 = 到達時間 + 執行時間 
            doneJob.delay = 0 # 等待時間 = 0 
        else :
            doneJob.departure = nowTime + aJob.duration # 執行完時間 = 現在時間 + 執行時間
            doneJob.delay = nowTime - aJob.arrival # 等待時間 = 開始執行時間 - 到達時間 

        nowTime = doneJob.departure 
        self.doneJobs.append( doneJob ) 
        return nowTime
    
    
    # display all on screen
    def ShowAll( self ):
        self.__ShowAbortJobs() ;
        self.__ShowDoneJobs();
        self.__ShowStat();
    
    # write all as a file
    def PutAll( self ):
        fileName = "output" + fileTitleNum + ".txt"
        file = open( fileName, 'w' )
        print( "\t[Abort Jobs]\n\tOID\tAbort\tDelay", file = file )
        for i in range( 0, len(self.abortJobs) ):
            print( "[", i+1, "]", self.abortJobs[i].OID, "\t", self.abortJobs[i].abort, "\t", self.abortJobs[i].delay, file = file )
        print( "\t[Jobs Done]\n\tOID\tDeparture\tDelay", file = file )
        for i in range( 0, len(self.doneJobs) ):
            print( "[", i+1, "]", self.doneJobs[i].OID, "\t", self.doneJobs[i].departure, "\t", self.doneJobs[i].delay, file = file )
         
        print( "[Average Delay]\t", self.avgDelay, " ms\n[Success Rate]\t", self.successRate, " %", file = file )
        
        file.close()
```

```python
class Simulation:
    # constructor 
    def __init__( self, jobList ):
        self.jobs = jobList # a list of jobs
        self.answers = AnsList() # a set of answers
        self.allQ = JobQueue(3) # a set of queue
    
    # deal a job in queues
    def DealQ( self, nowTime ):
        job = JobType()
        job = self.allQ.DeQueue() # 刪除並取出
        if ( nowTime + job.duration <= job.timeout ): # 現在時間 + 到達時間 <= TimeOut 
            nowTime = self.answers.AddDoneJob( job, nowTime )
        else :
            nowTime = self.answers.AddAbortJob( job, nowTime, True )
        return nowTime
    
    # finish the remaining jobs in queues
    def FinishQ( self, nowTime ): 
        while ( self.allQ.IsEmpty() == False ):
            nowTime = self.DealQ( nowTime )
        return nowTime
    
    
    # shortest queue first
    def SQF( self ):
        nowTime = 0 # 現在執行時間 
        aJob = JobType()
        while ( self.jobs.IsEmpty() == False ) : # { 還有工作沒處理
            aJob = self.jobs.NextJob() # 下一個
            
            if ( self.allQ.IsFull() == False ): # 還可以排隊 
                self.allQ.EnQueue( aJob ) # 排入佇列 
            elif ( aJob.arrival < nowTime ): # 還沒做就逾時
                nowTime = self.answers.AddAbortJob( aJob, nowTime, False )
            else: # 如果還沒逾時，先執行Queue
                nowTime = self.DealQ( nowTime ) # 處理Queue 
                self.allQ.EnQueue( aJob ) # 加入排隊 
                
        # } while
        nowTime = self.FinishQ( nowTime ) ; # 把剩下排隊的處理完
        self.answers.ShowAll() # 印在螢幕上 
        self.answers.PutAll() # 寫成檔案 
    
    
```
