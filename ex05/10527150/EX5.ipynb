```python
import os
import copy
import time
import sys
sys.setrecursionlimit(9000000) # 增加遞迴上限

############################################################ Sorting
## '畢業人數'以STRING存於List[X][8]
## 所以比較時要轉為int
## 以下每個排序在比較時皆是如此
############################################################ bubble sort
def bubble_sort( bubbleList ):
    timeStart = time.time() #記開始時間

    for i in range(0,len(bubbleList)-1):
        flag = False #紀錄是否已排序 以提前停止
        for j in range(0,len(bubbleList)-1-i):
            if int(bubbleList[j][8]) < int(bubbleList[j+1][8]): # 相鄰兩兩比較大小
                flag = True
                bubbleList[j],bubbleList[j+1] = bubbleList[j+1],bubbleList[j] # 交換
        if flag == False:
            break

    print("\nBubble sort:",int(round((time.time() - timeStart) * 1000)), "ms") #印出(當前時間-開始時間) *1000 並四捨五入 得到毫秒
    return bubbleList


############################################################ selection sort
def selection_sort( selectionList ):
    timeStart = time.time() #記開始時間

    n = len( selectionList )
    for i in range( 0, n ): # 從第一筆開始往後找最大值 最大值到最前後在從下一筆開始...
        j = i # j 用來記錄最大值的INDEX
        for k in range( i + 1, n ): # 用第一筆先當最大值 開始一個一個比較
            if int(selectionList[k][8]) >= int(selectionList[j][8]): # 若是找到新的最大值 更新j
                j = k
        selectionList[i], selectionList[j] = selectionList[j], selectionList[i] # 最大值跟最前面交換

    print("Selection sort:",int(round((time.time() - timeStart) * 1000)), "ms\n") #印出(當前時間-開始時間) *1000 並四捨五入 得到毫秒
    return selectionList


############################################################ quick Sort
def QuickSort( quickList ):
    timeStart = time.time() #記開始時間
    n = len( quickList )
    _QuickSort( quickList, 0, n - 1 )
    print("\nQuick sort:",int(round((time.time() - timeStart) * 1000)), "ms") #印出(當前時間-開始時間) *1000 並四捨五入 得到毫秒
    return quickList

def _QuickSort(quickList, left, right):
    if left >= right:
        return
    p = partition(quickList, left, right) # p為樞紐移到中間後的INDEX
    _QuickSort(quickList, left, p-1)  # 左側的遞迴 (左 ~ pivtor-1)
    _QuickSort(quickList, p+1, right) # 右側的遞迴 (pivtor+1 ~ 右)

def partition(quickList, left, right):
    pivot = quickList[left] #以首筆資料做為樞紐
    low = left + 1
    high = right

    while True:
        while low <= high and int(quickList[high][8]) <= int(pivot[8]): # 從右邊開始找，找比樞紐大的值
            high -=  1
        while low <= high and int(quickList[low][8]) >= int(pivot[8]):  # 從左邊開始找，找比樞紐小的值
            low +=  1
        if low <= high:  # 當左右沒有相遇時，互換值且外迴圈繼續
            quickList[low], quickList[high] = quickList[high], quickList[low]
        else:
            break
    quickList[left], quickList[high] = quickList[high], quickList[left] # 把樞紐交換到中間(小於右側大於左側)
    return high


############################################################ merge Sort
def MergeSort( mergeList ):
    timeStart = time.time() #記開始時間
    n = len( mergeList )
    _MergeSort( mergeList, 0, n - 1 )
    print("Merge sort:",int(round((time.time() - timeStart) * 1000)), "ms\n") #印出(當前時間-開始時間) *1000 並四捨五入 得到毫秒
    return mergeList

def _MergeSort( mergeList, left, right ):
    if left >= right:
        return

    mid = ( left + right ) // 2 #中間值
    _MergeSort( mergeList, left, mid )      # 左側的遞迴 (左 ~ 中)
    _MergeSort( mergeList, mid + 1, right ) # 右側的遞迴 (中 + 1 ~ 右)
    Merge( mergeList, left, mid, right )    # 合併左右

def Merge( mergeList, left, mid, right ):
    numL = mid - left + 1 #計算左數量
    numR = right - mid    #計算右數量
    L = [] #放左側numL筆資料
    R = [] #放右側numR筆資料

    for i in range( numL ):
        L.append( mergeList[left + i] )
    for j in range( numR ):
        R.append( mergeList[mid + j + 1] )

    i = j = 0 # i 紀錄左側INDEX ，j 紀錄右側INDEX
    for k in range( left, right + 1 ): # left ~ right + 1
        if i < numL and j < numR: #比較左右側(L[i]，R[j]) 較大者放入mergeList[k]
            if int(L[i][8]) >= int(R[j][8]):
                mergeList[k] = L[i]
                i += 1
            else:
                mergeList[k] = R[j]
                j += 1
        elif i < numL and j >= numR: #剩左側 放入mergeList[k]
            mergeList[k] = L[i]
            i += 1
        else:                        #剩右側 放入mergeList[k]
            mergeList[k] = R[j]
            j += 1

############################################################# file process & interface
def readFile( SchoolList ):

    while True :

        filenum = input("\ninput file name ...[0]Quit) : ")
        if filenum == "0":
            return filenum
        filename = 'input'+ filenum +'.txt'

        if os.path.isfile(filename) :
            with open( filename , 'r' ) as  f:

                for line in f: # 一行一行以STRING讀入
                    line = line.split('\t', ) # 以tab分割存到LIST
                    SchoolList.append(line)   # append 分隔後的LIST到SchoolList(二維)

            for i in  range( 3 ): # 去除前三行
                SchoolList.pop(0)

            return filenum # 回傳檔案編號

        else :
            print("檔案不存在。")


def WriteFile( fnum , outList ) :
    OutputFileName = fnum +".txt" # 檔名
    OutputFile = open(OutputFileName, "w", encoding='UTF-8') # 以寫檔模式開啟檔案;不存在則建立

    for i in range(len(outList)):
        for j in range(len(outList[i])):
            if (j == 10):                                        # 每個學校資料的"體系別"在讀檔時連換行符號一起讀入了
                OutputFile.write(outList[i][j].replace('\n','')) # 但最後一筆學校資料的"體系別"不一定會有換行符號
                OutputFile.write('\n')                           # 因此不論有沒有'\n'皆先去掉，在統一換行
            else :
                OutputFile.write(outList[i][j])
                OutputFile.write('\t') # 每一項以TAB分開
    OutputFile.close()

    
def PrintAll( outList ) : # 測試用
    for i in range(len(outList)):
        for j in range(len(outList[i])):
            if (j == 10):                                        
                print(outList[i][j].replace('\n',''), end='') 
                print()                           
            else :
                print(outList[i][j], end='')
                print('\t', end='') 


def InterFace():
    print("*********** Sorting Algorithms *********")
    print("* 0. Quit                              *")
    print("* 1. Selection Sort vs. Bubble Sort    *")
    print("* 2. Merge Sort vs. Quick Sort         *")
    print("****************************************")
    command = input("\nInput a command( 0, 1, 2 ):  ")
    return command # 回傳使用者輸入之COM


if __name__ == '__main__':

    quit = False #主程式迴圈停止條件
    com = 0 #使用者input
    SchoolList = [] #存原始檔案

    while (quit != True) :

        SchoolList.clear() #初始化LIST
        com = InterFace( ) #叫出使用者介面並回傳com

        if com == '0' :
            quit = True # 結束迴圈(程式)
            print("\nThanks for using !")

        elif com == '1' :

            fileNum = readFile( SchoolList ) #讀檔並存到SchoolList並回傳檔案編號

            if fileNum != "0" :

                BubbleList = copy.deepcopy(SchoolList)    ### 用deepcopy 分出兩個排列專用的LIST
                SelectionList = copy.deepcopy(SchoolList) #   因在python中 若是使用 a = b 賦值將會使a指向b的記憶體位置
                                                          #   所以若是之後進行 b.append() 等等的操作時也將會動到 a 這個LIST
                                                          #   而若是使用 a = copy.copy(b) 只會拷貝内部的元素的引用而已
                                                          #   因此若是拷貝二維LIST時後做 b[i].append() 時也會動到a[i]這個LIST
                                                          #                         進行 b.append() 則不會動到 a 這個LIST

                BubbleList = bubble_sort( BubbleList )
                SelectionList = selection_sort( SelectionList )

                WriteFile( "bubble_sort"+  fileNum ,BubbleList )    # 將結果輸出
                WriteFile( "select_sort"+  fileNum ,SelectionList ) # 將結果輸出

        elif com == '2' :


            fileNum = readFile( SchoolList ) #讀檔並存到SchoolList並回傳檔案編號

            if fileNum != "0" :
                QuickList = copy.deepcopy(SchoolList)     # 用deepcopy 分出兩個排列專用的LIST
                MergeList = copy.deepcopy(SchoolList)     # 原因同上

                QuickList = QuickSort(QuickList)
                MergeList = MergeSort(MergeList)

                WriteFile( "quick_sort"+  fileNum ,QuickList )      # 將結果輸出
                WriteFile( "merge_sort"+  fileNum ,MergeList )      # 將結果輸出

        else:
            print("\nERROR command!\n")


```
