```python
class node:
    def __init__(self, schoolcode, schoolname, majorcode, majorname, time, level, studentNum, teacherNum, graduated, city, department):
        self.schoolcode =schoolcode
        self.schoolname =schoolname
        self.majorcode =majorcode
        self.majorname =majorname
        self.time =time
        self.level =level
        self.studentNum =studentNum
        self.teacherNum =teacherNum
        self.graduated =graduated
        self.city =city
        self.department =department
```

```python
class Sorting :
    def __init__(self) :
        self.test= []
        self.merge =[]
        self.quick =[]
        
        
        
    def Merge( self, test, start, mid, finish) : #mergesort
        
        first1 = start    #記錄前半段的範圍
        last1 = mid 
        first2 = mid + 1   # 記錄後半段的範圍
        last2 = finish  
        temp = Sorting().test
        while first1 <= last1 and first2 <= last2 :   #當前半或後半還有數值時，持續比對
            if int(test[first1].graduated) >= test[first2].graduated :
                temp.append(test[first1])
                first1 += 1
            else:
                temp.append(test[first2])
                first2 += 1

        #當左右邊其中一邊比對完後，把剩下的資料直接放入temp，看是哪一邊還沒放完
        while first1 <= last1 :
            temp.append(test[first1])
            first1 += 1

        while first2 <= last2 :
            temp.append(test[first2])
            first2 += 1

        index = start
        x = 0 
        while  index <= finish and x < len(test) :
            test[index] = temp[x]
            x += 1
            index += 1


    def MergeSort(self, test, start, finish):
                                                 #start與finish為範圍
        if  start < finish :                   # 表示目前的矩陣範圍是有效的
            mid = ( start + finish ) // 2;         # mid即是將矩陣對半分的index
            self.MergeSort( test, start, mid )    # 繼續divide矩陣的前半段subarray
            self.MergeSort( test, mid+1, finish )    # 繼續divide矩陣的後半段subarray
            self.Merge( test, start, mid, finish )      # 將兩個subarray做比較, 並合併出排序

        return test



                           

    def Partition(self, test, start, finish ):
        x = start  #x是紀錄比較大的數字的尾端
        pivot = test[start].graduated #將首筆資料設為pivot
        y = start + 1
        while y < finish :
            if test[y].graduated > pivot :  #檢查除了pivot(首筆資料)以外的數字，y是比較小的數字尾端
                x += 1
                test[x], test[y] = test[y], test[x] # 當y的資料比pivot大，交換x+1跟y的位置(放到x的排序裡面)
            y += 1
        test[x], test[start] = test[start], test[x] #將pivot放在x+1的位置(正確位置)

        return x

    def QuickSort(self, test, start, finish):
        if start < finish :
            pivot = self.Partition(test, start, finish)
            self.QuickSort(test, start, pivot - 1)
            self.QuickSort(test, pivot + 1, finish)

        return test

```
